```python
# to check the intensity i.e. positivity,negativity,neutrality of tweets
sid = SentimentIntensityAnalyzer()

listy=[]

#to get the polarity scores using SentimentItensityAnalyser
for index, row in data.iterrows():
  ss = sid.polarity_scores(row["Tweets"])
  listy.append(ss)
    
#to store the values in the list listy in form of columns
se = pd.Series(listy)

#add a new column named polarity in the data table
data['polarity'] = se.values

display(data.head(100))
```

```python
#store the different polarities in lists
negative=[]
positive=[]
neutral=[]
compound=[]
overall=[]

#list contains dictionaries that are to be accessed
for i in listy:
    if 'neg' in i:
        negative.append(i['neg'])
    if 'pos' in i:
        positive.append(i['pos'])
    if 'neu' in i:
        neutral.append(i['neu'])
    if 'compound' in i:
        compound.append(i['compound'])

#to get the overall polarity using compound         
for i in compound:
    if i>=0.05:
        overall.append('positive')
    elif i<=-0.05:
        overall.append('negative')
    else:
        overall.append('neutral')
    
print("Negative polarities:",negative)
print("Positive polarities:",positive)
print("Neutral polarities:",neutral)
print("Compound polarities:",compound)
print("Overall polarities:",overall)

data['overall polarity']=overall
print(data.head(10))
```

```python
#to get the total sum of polarities of all tweets
negative_sum=sum(negative)
positive_sum=sum(positive)
neutral_sum=sum(neutral)
compound_sum=sum(compound)
print("negative sum",negative_sum)
print("positive sum",positive_sum)
print("neutral sum",neutral_sum)
```

## ANALYSIS ON PIE CHART

```python
#import library to plot
from matplotlib import pyplot as plt
activities=['negative','positive','neutral']
slices=[negative_sum,positive_sum,neutral_sum]
cols=['g','r','y']
plt.pie(slices,labels=activities,colors=cols,startangle=90,shadow='True',explode=(0,0.3,0.4),autopct='%1.1f%%')
plt.show()
```
